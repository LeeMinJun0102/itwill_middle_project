## (1) total_road_kill <---> eco_road

```python
df_melt = df[['total_rk', 'national_rk', 'highway_rk', 'white_rk']].melt(var_name='variable', value_name='value')

plt.figure(figsize=(16, 8))
ax = sns.countplot(data=df_melt, x='value', hue='variable', color='red', edgecolor="white")

for container in ax.containers:
    ax.bar_label(container)

plt.grid()
plt.title('roadkill by road')
# plt.savefig("roadkill by road.png", dpi=300, bbox_inches='tight')
# files.download("roadkill by road.png")

plt.show()
```

```python
df_melt_over = df_melt[df_melt['value'] >= 1]

plt.figure(figsize=(16, 8))
ax = sns.countplot(data=df_melt_over, x='value', hue='variable', color='red', edgecolor="white")

for container in ax.containers:
    ax.bar_label(container)

plt.grid()
plt.title('roadkill by road')
# plt.savefig("roadkill by road zoom.png", dpi=300, bbox_inches='tight')
# files.download("roadkill by road zoom.png")
plt.show()                #확대한거 # 빨간게 고속도론데 고속도로 설명 원
```

## dist_wildlife_corridor

로드킬과 생태통로의 거리를 보면 2.5km 이내의 분포가 가장 많다

```python
ax = sns.histplot(data=df, x='dist_to_eco_m', color='lightblue', edgecolor="black")

for con in ax.containers:
    ax.bar_label(con)

plt.title('distance between rk and corridor')
plt.grid()
# plt.savefig("dist_to_eco_m_count.png", dpi=300, bbox_inches='tight')
# files.download("dist_to_eco_m_count.png")
plt.show()
```

```python
# 거리 대략 3000 이상
df2 = df.sort_values(by='dist_to_eco_m', ascending=False).head(145)

ax = sns.histplot(data=df2, x='dist_to_eco_m', color='lightblue', edgecolor="black")

for con in ax.containers:
    ax.bar_label(con)

plt.title('distance over 3000 between rk and corridor')
plt.grid()
# plt.savefig("dist_to_eco_m_count_zoom.png", dpi=300, bbox_inches='tight')
# files.download("dist_to_eco_m_count_zoom.png")
plt.show()          # 확대한거임
```
